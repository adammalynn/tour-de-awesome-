```python
# Build the endpoint URL
# use string formating syntax on string url
target_url2 = 'https://maps.googleapis.com/maps/api/geocode/json?address=%s&key='+config.gkey

list_2 = []
for item in dataList:
    
    call=requests.get(target_url2%item).json()
    lat_lng=[call['results'][0]['formatted_address'],
             call['results'][0]['geometry']['location']['lat'],
             call['results'][0]['geometry']['location']['lng']
            ]
    list_2.append(lat_lng)
list_2
```

### Join and query

```python
# Name of the CSV file
file = "resources/race_starters.csv"
# Use the correct encoding to read the CSV in pandas
starters_df = pd.read_csv(file, encoding="ISO-8859-1")
starters_df.head()
```

```python
# Reference multiple columns within a DataFrame
starters_data = starters_df[["rider_id","rider_name","rider_country"]]
starters_data.head()
```

```python
starters_data.count()
```

```python
# Name of the CSV file
file = "resources/race_results.csv"
# Use the correct encoding to read the CSV in pandas
finishers_df = pd.read_csv(file, encoding="ISO-8859-1")
finishers_df.head()
```

```python
# Reference multiple columns within a DataFrame
finishers_data = finishers_df[["rider_id","stage_id","race_result_type_id","ranking", "rider_time"]]
finishers_data.head()
```

```python
# Collect only those columns filtered by race_result_type_id is 2
race_result_typeid_2 = finishers_df.loc[finishers_df["race_result_type_id"] == 2]
race_result_typeid_2.head()
```

```python
stage_id_21 = race_result_typeid_2.loc[race_result_typeid_2["stage_id"] == 21]
stage_id_21.head()
```

```python
# Reference multiple columns within a DataFrame
stage21_df = stage_id_21[["rider_id","ranking", "rider_time"]]
stage21_df.head()
```

```python
stage21_df.count()
```

```python
country_data.head()
```

```python
country_data.count()
```

```python
# Merge two dataframes using an outer join
first_merge = pd.merge(starters_data, country_data, on="rider_country", how="outer")
first_merge.head()
```

```python
# Merge two dataframes using an outer join
second_merge = pd.merge(first_merge, stage21_df, on="rider_id", how="outer")
second_merge.head()
```
